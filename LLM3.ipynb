```python
from langchain.document_loaders import TextLoader

documents = TextLoader("C:/Users/chogu/OneDrive/바탕 화면/대학교/4학년/모바일 프로그래밍 실습/10주차/2/080413-main/Data/AI.txt").load()   # AI.txt 파일 위치 지정
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

# 문서의 내용을 청크로 분할
def split_docs(documents, chunk_size=1000, chunk_overlap=200):
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=chunk_size,
        chunk_overlap=chunk_overlap
    )
    docs = text_splitter.split_documents(documents)
    return docs

# docs 변수에 분할된 문서를 저장
docs = split_docs(documents)
```

```python
from langchain.schema import Document

docs = [
    Document(page_content="사람은 무엇인가? 인간은 사회적 동물이다."),
    Document(page_content="강아지는 귀엽다."),
    Document(page_content="AI는 머신러닝 기술을 기반으로 발전한다.")
]
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_chroma import Chroma

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

db = Chroma.from_documents(
    documents=docs,
    embedding=embeddings,
    persist_directory="./chroma_db"
)
```

```python
import os     # 환경변수(OS) 모듈을 가져와 API 사용에 필요한 라이브러리
os.environ["OPENAI_API_KEY"] = "sk"   # OpenAI 키 입력

from langchain.chat_models import ChatOpenAI
model_name = "gpt-4o-mini"   # gpt-3.5-turbo 모델 사용

llm = ChatOpenAI(model_name=model_name)

from langchain.chains.question_answering import load_qa_chain
chain = load_qa_chain(llm, chain_type="stuff", verbose=True)

# 질의를 입력하고 관련 문서를 검색한 후 답변을 생성. 미리 텍스트에 있는 내용을 질의에 입력한다.
query = "사람"
matching_docs = db.similarity_search(query)
answer = chain.run(input_documents=matching_docs, question=query)
answer
```
